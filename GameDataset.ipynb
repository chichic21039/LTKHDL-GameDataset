## 1.4 Dữ liệu được thu thập như thế nào
### 1.4.1 Phương pháp thu thập
• Dữ liệu được tác giả thu thập bằng phương pháp web scraping từ trang vgchartz.com. 
### 1.4.2 Đối tượng mục tiêu và phương pháp lấy mẫu:
• Đối tượng mục tiêu: là dữ liệu doanh số các tựa game trên toàn cầu, nhưng chủ yếu trung vào game bán lẻ(retail) vì VGCharts đã ngừng ước tính  tính doanh số kỹ thuật số từ 2018

• Phương pháp lấy mẫu: Ở đây tác giả không đề cập đến sẽ lấy các mẫu nào chỉ nói đến việc thu thập dữ liệu bằng một chương trình tự động, tức là lấy mẫu toàn bộ dữ liệu bao gồm tất cả các tựa game có sẵn trên VGCharts
### 1.4.3 Khoảng thời gian thu thập dữ liệu
Tác giả không đề cập đến thời gian cào dữ liệu nhưng trang web VGChartz được thành lập từ (2005) và từ năm 2018 trở đi thì VGChart ngừng ước tính doanh số phần mềm vì thị trường kỹ thuật số chiếm tỷ trọng lớn,khó đo lường và dữ liệu chỉ có trường "date" là ngày phát hành game và không có ngày công bố trên VGChart, dữ liệu ngày phát hành game với khoảng vùng 01-01-1970 đến 31-12-2024
### 1.4.4 Hạn chế và bias dữ liệu
Do sau năm 2018 VGChart ngừng ước tính doanh số phần mềm nên dữ liệu chủ yếu phản ánh doanh số bán lẻ vật lý, nên có thể dữ liệu về doanh số phần mềm sẽ bị thiếu hoặc ít và phần lớn dữ liệu doanh số cũng chỉ là ước tính từ VGChart chỉ một số ít là từ nhà phát hành nên cũng có thể có rủi ro về ước tính
