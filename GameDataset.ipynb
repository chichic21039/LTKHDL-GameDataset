# 1. Data collection

1. Đề tài của bộ dữ liệu <br>

Chúng em chọn dataset VGChartz (Video Game Sales) vì:

- Chủ đề cực kỳ hấp dẫn và gần gũi: ngành game hiện lớn hơn cả phim và nhạc cộng lại, mọi thành viên trong nhóm đều đam mê và muốn khám phá lý do vì sao một số game thành “huyền thoại” về doanh số.
- Nguồn dữ liệu uy tín: VGChartz là một trong những trang theo dõi doanh số game nổi tiếng nhất thế giới, được giới chuyên môn tin dùng nhiều năm.
- Rất giàu thông tin: hơn 55.000 tựa game với doanh số theo khu vực, nền tảng, nhà phát hành, điểm đánh giá – đủ để làm từ phân tích cơ bản đến mô hình dự đoán doanh số.
- Phù hợp mọi trình độ
- Dữ liệu đã khá sạch và dễ tiếp cận trên Kaggle.

Dataset VGChartz cho phép trả lời rất nhiều câu hỏi thú vị như:

- Năm nào là đỉnh cao của ngành game về số lượng và doanh số?
- Tháng nào trong năm là mùa phát hành game nóng nhất và vì sao?
- Thị trường Nhật Bản khác biệt thế nào so với Bắc Mỹ và châu Âu?
- Nền tảng và nhà phát hành nào thống trị ngành qua các thời kỳ?
- Điểm số của giới phê bình và người chơi có thực sự quyết định doanh số không?
- Dataset này không chỉ cho chúng em thấy ai bán chạy nhất, mà còn giúp hiểu rõ công thức thành công thực sự của một tựa game trong ngành công nghiệp hàng trăm tỷ đô la này.
