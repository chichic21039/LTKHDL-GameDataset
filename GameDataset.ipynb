## 2.2 Phân tích cột dạng số

```python
numeric_cols = ["shipped", "total", "america", "europe", "japan", "other", "vgc", "critic", "user"]
print("Numeric:", numeric_cols)

def detect_distribution(data):
    skew = data.skew()
    if abs(skew) < 0.5:
        return "Approximately Normal"
    elif skew > 0:
        return "Right-Skewed"
    else:
        return "Left-Skewed"

for col in numeric_cols:
    print(f"\n===== COLUMN: {col.upper()} =====")
    
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
    mean = df[col].mean()
    median = df[col].median()
    std = df[col].std()
    min_val = df[col].min()
    max_val = df[col].max()
    
    print(f"Mean   : {mean}")
    print(f"Median : {median}")
    print(f"Std Dev: {std}")
    print(f"Min    : {min_val}")
    print(f"Max    : {max_val}")

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col].count()
    print(f"Outliers (IQR method): {outliers}")
    
    # Dạng phân phối
    dist_type = detect_distribution(df[col].dropna())
    print(f"Distribution: {dist_type}")
    
    # % Missing values
    missing_pct = df[col].isna().mean() * 100
    print(f"Missing values: {missing_pct:.2f}%")
    
    # Placeholder bất thường
    placeholder = df[col].isin([0, -1, 999]).sum()
    print(f"Suspicious placeholders (0, -1, 999): {placeholder}")
    
    # Giá trị âm (bất hợp lý)
    negative_values = (df[col] < 0).sum()
    print(f"Negative values: {negative_values}")
    
    # Vẽ biểu đồ
    plt.figure(figsize=(15, 4))
    
    # Histogram
    plt.subplot(1, 3, 1)
    plt.hist(df[col].dropna(), bins=30)
    plt.title(f"{col} - Histogram")
    
    # Boxplot
    plt.subplot(1, 3, 2)
    plt.boxplot(df[col].dropna(), vert=False)
    plt.title(f"{col} - Boxplot")

    #Density Plot
    plt.subplot(1, 3, 3)
    df[col].dropna().plot(kind="density", title=f"{col} - Density Plot")
    
    plt.tight_layout()
    plt.show()
```

## 2.3 Phân tích cột dạng phân loại

Chúng ta sẽ phân tích các cột phân loại (categorical) trong dataset:
- `name`, `platform`, `publisher`, `developer`


```python
categorical_cols = ["name", "platform", "publisher", "developer"]
print("Categorical:", categorical_cols)
```

#### Phân bố giá trị (Value Distribution)

Kiểm tra cho mỗi cột:
- Số lượng giá trị duy nhất
- Top 5 giá trị phổ biến nhất kèm tỷ lệ %
- Biểu đồ tần suất (count plot)
- Xem phân bố cân bằng hay lệch


```python
for col in categorical_cols:
    print(f"=== Cột: {col} ===")
    
    # Số giá trị duy nhất
    n_unique = df[col].nunique(dropna=True)
    print(f"Số giá trị duy nhất: {n_unique}")
    
    # Top 5 giá trị phổ biến + % so với tổng
    top5 = df[col].value_counts().head(5)
    top5_pct = df[col].value_counts(normalize=True).head(5) * 100
    print("Top 5 giá trị phổ biến (số lượng và %):")
    for val, cnt, pct in zip(top5.index, top5.values, top5_pct.values):
        print(f"- {val}: {cnt} lần, chiếm {pct:.2f}% tổng số")
    
    # Biểu đồ tần suất (top 10)
    plt.figure(figsize=(8,4))
    top10 = df[col].value_counts().head(10)
    sns.barplot(x=top10.values, y=top10.index)
    plt.xlabel("Số lượng")
    plt.ylabel(col)
    plt.title(f"Top 10 giá trị phổ biến cho cột {col}")
    plt.show()
    
    print("\n")
```

Nhận xét: 
- Cột "name" phân bố khá đều.
- Cột "platform" phân bố đều, với nền tảng PC chiếm số lượng lớn nhất (khoảng 20% tổng số), số lượng nền tảng khá ít so với số lượng tựa game, nhà phát hành, nhà phát triển.

```python
for col in categorical_cols:
    print(f"=== Cột: {col} ===")
    
    # Tỷ lệ missing
    missing_pct = df[col].isna().mean() * 100
    print(f"Tỷ lệ giá trị thiếu: {missing_pct:.2f}%")
    
    # Các giá trị duy nhất (sample) để kiểm tra inconsistencies/typos
    unique_vals = df[col].unique()
    print(f"Một số giá trị duy nhất (max 20 giá trị): {unique_vals[:20]}")
    
    # Các category ít xuất hiện (<5 lần)
    value_counts = df[col].value_counts()
    rare_categories = value_counts[value_counts < 5]
    
    if len(rare_categories) > 0:
        # Tỷ lệ % số category hiếm so với tổng số category duy nhất
        rare_pct = len(rare_categories) / len(value_counts) * 100
        print(f"Các category ít xuất hiện (<5 lần):")
        print(rare_categories)
        print(f"Tỷ lệ các category hiếm so với tổng số category: {rare_pct:.2f}%")
    else:
        print("Không có category ít xuất hiện (<5 lần).")
    
    print("\n")

```

```python


# Ép kiểu datetime; chỉnh format/dayfirst nếu cần
df['date'] = pd.to_datetime(df['date'], errors='coerce')

print("Sau to_datetime, số giá trị NaN:", df['date'].isna().sum())

# Điền nan bằng median theo platform
median_by_platform = df.groupby('platform')['date'].transform('median')
df['date'] = df['date'].fillna(median_by_platform)

# Nếu vẫn còn nan (platform cũng toàn nan), thì tiếp tục điền bằng median theo publisher
still_nan_idx = df['date'].isna()
if still_nan_idx.any():
    median_by_publisher = df.groupby('publisher')['date'].transform('median')
    df.loc[still_nan_idx, 'date'] = median_by_publisher[still_nan_idx]

# Kiểm tra
missing_after = df['date'].isna().sum()
print("Số NaN còn lại:", missing_after)
if missing_after:
    # Xem nhóm nào vẫn NaT (nghĩa là nhóm không còn giá trị hợp lệ)
    still_nat = df[df['date'].isna()][['platform', 'publisher']].value_counts()
    print("Platform/Publisher vẫn NaT:", still_nat)
else:
    print("Đã xử lý hết NaN.")
```

## 2.5 Tương quan và mối quan hệ giữa các thuộc tính

### 2.5.1 Các pattern sơ bộ
- Lập ma trận tương quan cho các cột numeric
- Tạo heatmap tương quan
- Xác định các cặp thuộc tính tương quan mạnh

```python
corr_matrix = df[numeric_cols].corr()
print(corr_matrix)

# Heatmap trực quan
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Heatmap tương quan các biến numeric")
plt.show()

# Tìm các cặp biến tương quan mạnh (>|0.7|)
threshold = 0.7
strong_corr = []

for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i,j]) >= threshold:
            strong_corr.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i,j]))

print("Các cặp biến numeric có tương quan mạnh (>|0.7|):")
for col1, col2, val in strong_corr:
    print(f"- {col1} và {col2}: {val:.2f}")
```

### 2.5.2 Phân tích chéo dữ liệu (Cross-tabulations)
- Phân tích các cột phân loại × phân loại (Categorical × Categorical) bằng bảng tần số.
- Phân tích các cột số × phân loại (Numerical × Categorical) bằng thống kê nhóm.


#### Phân tích Categorical × Categorical

Tạo bảng tần số cho các cặp cột phân loại quan trọng, ví dụ:
- `platform` × `publisher`
- `platform` × `developer`


```python
cat_pairs = [
    ("publisher", "platform"),
    ("developer", "platform"),
    ("publisher", "developer"),
]

topN = 10

for cat1, cat2 in cat_pairs:
    print(f"\n=== {cat1} × {cat2} (Top {topN}) ===")
    
    top1 = df[cat1].value_counts().head(topN).index
    top2 = df[cat2].value_counts().head(topN).index
    
    sub = df[df[cat1].isin(top1) & df[cat2].isin(top2)]
    
    ct = pd.crosstab(sub[cat1], sub[cat2])
    display(ct)
```

#### Phân tích các cột Numerical x Categorical

```python
pairs = [
    ('publisher', 'critic'),
    ('publisher', 'user'),
    ('publisher', 'total'),
    ('developer', 'critic'),
    ('developer', 'user'),
    ('developer', 'total'),
    ('platform', 'critic'),
    ('platform', 'user'),
    ('platform', 'total'),
    ('publisher', 'america')
]


for cat, num in pairs:
    if num not in df.columns:
        print(f"Bỏ qua cặp {cat} - {num} (không tồn tại)")
        continue
    # chọn top 10 giá trị cate phổ biến nhất
    top_vals = df[cat].value_counts().head(10).index
    sub = df[df[cat].isin(top_vals)]

    plt.figure(figsize=(12,6))
    sns.barplot(data=sub, x=cat, y=num, estimator='mean', errorbar=None)
    plt.title(f"{num} trung bình theo {cat} (Top 10 {cat})")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```
