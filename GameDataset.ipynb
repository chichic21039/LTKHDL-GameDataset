### 2.1.2 Tính toàn vẹn dữ liệu

```python
dup_count = df.duplicated().sum()
print(f"Số dòng bị lặp lại trong dataset là: {dup_count}")

df = df.drop_duplicates()
print("Đã loại bỏ các dòng trùng lặp khỏi dataset.")
```

Ở đây em chọn loại bỏ các dòng bị trùng lặp vì các dữ liệu này là có 3 thuộc tính để xác định nó là duy nhất là một tên game,ngày phát hành,nền tảng,duy nhất nên không phải ngẫu nhiên có thể trùng lặp nên sự trùng lặp dòng này là do sơ suất không phải dữ liệu ngẫu nhiên trùng lặp

```python
empty_cols = df.columns[df.isnull().all()].tolist()
if empty_cols:
    print("Các cột trống hoàn toàn:", empty_cols)
else:
    print("Không có cột nào trống hoàn toàn.")
```

### 2.1.3 Danh mục cột


 Ý nghĩa của từng cột trong dataset:
 - name: tên của trò chơi điện tử.
 - date: ngày phát hành trò chơi điện tử.
 - platform: nền tảng chơi game (All – tất cả nền tảng, Series – toàn bộ series game).
 - publisher: nhà phát hành.
 - developer: nhà phát triển.
 - shipped: số lượng bản đã xuất xưởng (chỉ áp dụng với các bản ghi có platform là All hoặc Series).
 - total: tổng số bản đã bán ra (tính theo triệu bản).
 - america: số bản bán tại khu vực Châu Mỹ (triệu bản).
 - europe: số bản bán tại khu vực Châu Âu (triệu bản).
 - japan: số bản bán tại Nhật Bản (triệu bản).
 - other: số bản bán ở các khu vực khác trên thế giới.
 - vgc: điểm đánh giá từ VGChartz.com.
 - critic: đánh giá từ giới phê bình.
 - user: đánh giá từ người dùng.

 Một số cột tiềm năng có thể phân tích gồm:
 - date: giúp phân tích xu hướng phát hành game theo thời gian, xác định các mốc tăng trưởng, mùa vụ,...
 - platform: nghiên cứu sự khác biệt về doanh số, thị phần hoặc điểm đánh giá giữa các nền tảng hoặc series.
 - publisher & developer: có thể phân tích mức độ thành công của từng nhà phát hành/nhà phát triển.
 - critic & user: cho phép so sánh giữa đánh giá chuyên môn và đánh giá từ người chơi, phân tích các yếu tố ảnh hưởng đến điểm số.

Không có cột nào nên được drop vì từng cột sẽ có ý nghĩa để phân tích khác nhau tùy theo khía cạnh
