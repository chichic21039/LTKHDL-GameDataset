## 2.2 Phân tích cột dạng số

```python
numeric_cols = [
    "shipped", "total", "america", "europe", "japan", 
    "other", "vgc", "critic", "user",
    "date"
]

print("Numeric:", numeric_cols)

def detect_distribution(data):
    skew = data.skew()
    if abs(skew) < 0.5:
        return "Approximately Normal"
    elif skew > 0:
        return "Right-Skewed"
    else:
        return "Left-Skewed"

for col in numeric_cols:
    print(f"\n===== COLUMN: {col.upper()} =====")

    # ===== SPECIAL CASE: DATE =====
    if col == "date":
        series = df["date"].dt.year
        series_name = "year (from date)"
    else:
        series = pd.to_numeric(df[col], errors="coerce")
        series_name = col

    mean = series.mean()
    median = series.median()
    std = series.std()
    min_val = series.min()
    max_val = series.max()

    print(f"Mean   : {mean}")
    print(f"Median : {median}")
    print(f"Std Dev: {std}")
    print(f"Min    : {min_val}")
    print(f"Max    : {max_val}")

    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = series[(series < lower_bound) | (series > upper_bound)].count()
    print(f"Outliers (IQR method): {outliers}")

    dist_type = detect_distribution(series.dropna())
    print(f"Distribution: {dist_type}")

    missing_pct = series.isna().mean() * 100
    print(f"Missing values: {missing_pct:.2f}%")

    negative_values = (series < 0).sum()
    print(f"Negative values: {negative_values}")

    # ===== VISUALIZATION =====
    plt.figure(figsize=(10, 3))

    plt.subplot(1, 2, 1)
    plt.hist(series.dropna(), bins=30)
    plt.title(f"{series_name} - Histogram")

    plt.subplot(1, 2, 2)
    plt.boxplot(series.dropna(), vert=False)
    plt.title(f"{series_name} - Boxplot")

    plt.tight_layout()
    plt.show()
```

Chúng ta sẽ phân tích các cột phân loại (categorical) trong dataset:

```python
categorical_cols = ["name", "platform", "publisher", "developer"]
print("Categorical:", categorical_cols)
```

#### Phân bố giá trị (Value Distribution)

Kiểm tra cho mỗi cột:
- Số lượng giá trị duy nhất
- Top 5 giá trị phổ biến nhất kèm tỷ lệ %
- Biểu đồ tần suất (count plot)
- Xem phân bố cân bằng hay lệch


```python
for col in categorical_cols:
    print(f"=== Cột: {col} ===")
    
    # Số giá trị duy nhất
    n_unique = df[col].nunique(dropna=True)
    print(f"Số giá trị duy nhất: {n_unique}")
    
    # Top 15 giá trị phổ biến + % so với tổng
    top15 = df[col].value_counts().head(15)
    top15_pct = df[col].value_counts(normalize=True).head(15) * 100
    print("Top 15 giá trị phổ biến (số lượng và %):")
    for val, cnt, pct in zip(top15.index, top15.values, top15_pct.values):
        print(f"- {val}: {cnt} lần, chiếm {pct:.2f}% tổng số")
    
    # Biểu đồ tần suất (top 15)
    plt.figure(figsize=(8,4))
    top15 = df[col].value_counts().head(15)
    sns.barplot(x=top15.values, y=top15.index)
    plt.xlabel("Số lượng")
    plt.ylabel(col)
    plt.title(f"Top 15 giá trị phổ biến cho cột {col}")
    plt.show()
    
    print("\n")
```

Nhận xét: 
- Cột "name" phân bố khá đều.
- Cột "platform" phân bố đều, với nền tảng PC chiếm số lượng lớn nhất (khoảng 20% tổng số), số lượng nền tảng khá ít so với số lượng tựa game, nhà phát hành, nhà phát triển.

```python
for col in categorical_cols:
    print(f"=== Cột: {col} ===")
    
    # Tỷ lệ missing
    missing_pct = df[col].isna().mean() * 100
    print(f"Tỷ lệ giá trị thiếu: {missing_pct:.2f}%")
    
    # Các giá trị duy nhất (sample) để kiểm tra inconsistencies/typos
    unique_vals = df[col].unique()
    print(f"Một số giá trị duy nhất (max 20 giá trị): {unique_vals[:20]}")
    
    # Các category ít xuất hiện (<5 lần)
    value_counts = df[col].value_counts()
    rare_categories = value_counts[value_counts < 5]
    
    if len(rare_categories) > 0:
        # Tỷ lệ % số category hiếm so với tổng số category duy nhất
        rare_pct = len(rare_categories) / len(value_counts) * 100
        print(f"Các category ít xuất hiện (<5 lần):")
        print(rare_categories)
        print(f"Tỷ lệ các category hiếm so với tổng số category: {rare_pct:.2f}%")
    else:
        print("Không có category ít xuất hiện (<5 lần).")
    
    print("\n")

```

### 2.5.2 Phân tích chéo dữ liệu (Cross-tabulations)
- Phân tích các cột phân loại × phân loại (Categorical × Categorical) bằng bảng tần số.
- Phân tích các cột số × phân loại (Numerical × Categorical) bằng thống kê nhóm.


#### Phân tích Categorical × Categorical

Tạo bảng tần số cho các cặp cột phân loại quan trọng, ví dụ:
- `platform` × `publisher`
- `platform` × `developer`
- `publisher` × `developer`


```python
cat_pairs = [
    ("publisher", "platform"),
    ("developer", "platform"),
    ("publisher", "developer"),
]

topN = 10

for cat1, cat2 in cat_pairs:
    print(f"\n=== {cat1} × {cat2} (Top {topN}) ===")
    
    top1 = df[cat1].value_counts().head(topN).index
    top2 = df[cat2].value_counts().head(topN).index
    
    sub = df[df[cat1].isin(top1) & df[cat2].isin(top2)]
    
    ct = pd.crosstab(sub[cat1], sub[cat2])
    display(ct)
```

Đây không phải lỗi ngẫu nhiên mà xuất phát từ thiết kế dataset, đặc biệt khi nền tảng mang giá trị All/Series hoặc không có ghi nhận bán lẻ theo khu vực. Điều này cho thấy việc sử dụng các biến doanh số cho phân tích hoặc mô hình cần được cân nhắc rất cẩn thận.

2. Các cột đánh giá (VGC, Critic, User) có chất lượng tốt nhưng thiếu dữ liệu nghiêm trọng

Mặc dù phân bố điểm khá “sạch” và logic, nhưng tỷ lệ NaN lên đến 89–99%, dẫn đến việc các cột này chỉ có giá trị phân tích hạn chế. Dữ liệu thiếu nhiều đến mức không thể áp dụng các phương pháp imputation thông thường mà không gây méo phân phối, do đó nhiều phân tích sẽ bị giới hạn hoặc phải loại bỏ hoàn toàn các cột này.

3. Dữ liệu danh mục (name, platform, publisher, developer) đa dạng nhưng phân bố không đều

Platform có phân bố rõ ràng với PC chiếm ưu thế. Ngược lại, publisher và developer chứa nhiều giá trị hiếm xuất hiện, phản ánh ngành công nghiệp game phân mảnh và có nhiều nhà phát triển nhỏ. Số lượng lớn giá trị “Unknown” cũng cho thấy vấn đề hụt thông tin trong quá trình thu thập dữ liệu.

4. Mối tương quan giữa các biến numeric chủ yếu là quan hệ hiển nhiên và mang tính cấu trúc:
- total tương quan rất mạnh với america, europe, other do công thức cộng lại.
- vgc và critic tương quan cao vì đều là thang đánh giá chất lượng game.
- Các biến còn lại có tương quan yếu → không có mẫu hành vi bất thường hoặc mối quan hệ tiềm ẩn nào nổi bật.
- Điều này cho thấy dataset không có pattern numeric phức tạp, và các câu hỏi chuyên sâu sẽ phải dựa vào phân tích theo thời gian hoặc theo nhóm danh mục thay vì tương quan numeric.

### 2.6.2 Red Flags (Các vấn đề nghiêm trọng)
1. Tỷ lệ missing cực cao và có tính hệ thống ở các cột doanh số
- Nhiều cột có 70–90% dữ liệu trống. Điều này khiến phần lớn các biến doanh số không phù hợp để dùng cho mô hình hóa hoặc phân tích thống kê chuẩn.
2. Dữ liệu đánh giá thiếu trầm trọng, gần như không thể xây dựng mô hình dự đoán dựa trên điểm số
- USER missing 99%, VGC missing 96%, CRITIC missing 89% → nếu cố gắng điền giá trị sẽ gây méo dữ liệu nặng. Đây là hạn chế lớn trong mọi nghiên cứu liên quan đến chất lượng game.
